Create dataframe without zero values called train_no_zeros

```python
train_no_zeros=train[train['z']!=0]
```

```python
df_of_zeros
```

Create five subdataframe from train datafrane contain the rows that carat atribute value in it between limits carat values in the zero values dataframe

I divide the three dataframe using carat value into the dataframes:

Dataframe 1 (train_a) : contain the rows with carat value greater than or equal to 0.69 and less than or equal to 0.73

Dataframe 2 (train_b) : contain the rows with carat value greater than or equal to 0.99 and less than or equal to 1.16

Dataframe 3 (train_c) : contain the rows with carat value greater than or equal to 1.48 and less than or equal to 1.52

Dataframe 4 (train_d) : contain the rows with carat value greater than or equal to 2.00 and less than or equal to 2.27

Dataframe 5 (train_e) : contain the rows with carat value greater than or equal to 2.75 and less than or equal to 2.85

```python
train_a=train[(train['carat']>=0.69)& (train['carat']<=0.73)]
train_b=train[(train['carat']>=0.99)& (train['carat']<=1.16)]
train_c=train[(train['carat']>=1.48)& (train['carat']<=1.52)]
train_d=train[(train['carat']>=2.00)& (train['carat']<=2.27)]
train_e=train[(train['carat']>=2.75)& (train['carat']<=2.85)]
```

```python
train_a
```

I divide the df_of_zeros dataframe which contain the zeros values into five dataframes using the carat value:

Dataframe 1 (df_zeros_1) : contain the row with carat value equal to 0.71

Dataframe 2 (df_zeros_2) : contain the rows with carat value greater than 0.99 and less than 1.16

Dataframe 3 (df_zeros_3) : contain the row with carat value equal to 1.50

Dataframe 4 (df_zeros_4) : contain the rows with carat value greater than 2.00 and less than 2.26

Dataframe 5 (df_zeros_5) : contain the row with carat value equal to 2.80

```python
df_zeros_1=df_of_zeros.head(1)
df_zeros_2=df_of_zeros[(df_of_zeros["carat"]>0.99)&(df_of_zeros["carat"]<1.16)]
df_zeros_3=df_of_zeros[(df_of_zeros["carat"]==1.50)]
df_zeros_4=df_of_zeros[(df_of_zeros["carat"]>2.00)&(df_of_zeros["carat"]<2.26)]
df_zeros_5=df_of_zeros[(df_of_zeros["carat"]==2.80)]
```

Replace the zeros values by the median using replace method

```python
df_zeros_1['z'].replace(0,train_a['z'].median(),inplace=True)
df_zeros_1['y'].replace(0,train_a['y'].median(),inplace=True)
df_zeros_1['x'].replace(0,train_a['x'].median(),inplace=True)

df_zeros_2['z'].replace(0,train_b['z'].median(),inplace=True)
df_zeros_2['y'].replace(0,train_b['y'].median(),inplace=True)
df_zeros_2['x'].replace(0,train_b['x'].median(),inplace=True)

df_zeros_3['z'].replace(0,train_c['z'].median(),inplace=True)

df_zeros_4['z'].replace(0,train_d['z'].median(),inplace=True)
df_zeros_4['y'].replace(0,train_d['y'].median(),inplace=True)
df_zeros_4['x'].replace(0,train_d['x'].median(),inplace=True)

df_zeros_5['z'].replace(0,train_e['z'].median(),inplace=True)
```

```python
df_zeros_1
```

Merge the three dataframes after replacing the zero values by the median into one dataframe called df_filled using concat() method

```python
df_filled=pd.concat([df_zeros_1,df_zeros_2,df_zeros_3,df_zeros_4,df_zeros_5])
df_filled
```

Merge the df_filled dataframe with train_no_zeros dataframe using concat() method

#4- Select and Train a Model

---




```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(train_full.values,dai_price)
```

```python
from sklearn.metrics import mean_squared_error

price_predictions = lin_reg.predict(train_full.values)
lin_mse = mean_squared_error(dai_price, price_predictions)
lin_rmse = np.sqrt(lin_mse)
print(lin_rmse)
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor(random_state = 42)
tree_reg.fit(train_full.values,dai_price)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, train_full.values, dai_price, scoring ="neg_mean_squared_error",cv = 15)
tree_rmse_scores = np.sqrt(-scores)
```

```python
print("Scores: ", tree_rmse_scores)
print("Mean: ", tree_rmse_scores.mean())
print("Standard Deviation: ", tree_rmse_scores.std())
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
forest_scores = cross_val_score(forest_reg,train_full.values, dai_price,scoring = "neg_mean_squared_error", cv = 10)
forest_rmse_scores = np.sqrt(-forest_scores)
```

```python
print("Scores: ", forest_rmse_scores)
print("Mean: ", forest_rmse_scores.mean())
print("Standard Deviation: ", forest_rmse_scores.std())
```

Using RandomizedSearchCV to search for the best combination of hyperparameter values for the RandomForestRegressor

```python
from sklearn.model_selection import RandomizedSearchCV

param_distribs = {
        'n_estimators': np.random.randint(750,1000,10),
        'max_features': np.random.randint(6, 8, 10),
    }

forest_reg = RandomForestRegressor(random_state=42)
rnd_search = RandomizedSearchCV(forest_reg, param_distributions=param_distribs,
                                n_iter=10, cv=10, scoring='neg_mean_squared_error', random_state=42)
rnd_search.fit(train_full.values, dai_price)
```

Use the reasults we got from  RandomizedSearchCV in GridSearchCV  

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

param_grid = [
              {'n_estimators': [750,778], 'max_features':[6]},
              {'bootstrap':[False], 'max_features':[2,3,4],'n_estimators':[3,10]}
]
forest_reg = RandomForestRegressor(random_state = 42)

grid_search = GridSearchCV(forest_reg,param_grid,cv = 10, scoring = 'neg_mean_squared_error',return_train_score = True)
grid_search.fit(train_full.values, dai_price)
```

Getting the best estimator

```python
grid_search.best_estimator_
```

Display the the evaluation scores for every estimator

```python
cvres = grid_search.cv_results_

for mean_score, params in zip(cvres["mean_test_score"],cvres["params"]):
  print(np.sqrt(-mean_score), params)
```

```python
feature_importances = grid_search.best_estimator_.feature_importances_
feature_importances
```

Display the relative importance scores of each attribute for making accurate predictions

```python
attributes =['carat','depth','table','length/width','z','0','1','2']
sorted(zip(feature_importances,attributes),reverse = True)
```

Make a predict on test data

```python
final_model =grid_search.best_estimator_
final_predictions =pd.Series(final_model.predict(test_full_))
my_predictions = pd.DataFrame({'Id':test_full['Id'],'price':final_predictions})
```
